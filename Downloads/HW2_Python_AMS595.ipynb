## 4. Functions

2a. (10 points) Write a function that determines if a number is prime.

````python
def isprime(n):
    # Finish this function
    # ...
    if n>=2: #Checking if the parameter is >= 2. 
        for y in range(2,n): #Checking the factors of the parameter
            if n % y == 0: #If the number the number is divisible by any number in the range, the number is not prime
                return False
    else:
        return False #return false if n<2 
    return True
````

````python
# test codes
print(isprime(2))
print(isprime(10))
print(isprime(17))
````

2b. (10 points) Write a function that finds all the factors of a number.

````python
def factorize(n):
    answer = []
    # Finish this function
    # ...
    for i in range(1,n+1): #Checking factors from 1 up to n
        if n % i==0: #Checks if the input parameter is divisible for intergers up to, and including, n
            answer.append(i) #Adds the factor to list
    return answer #After going through all elements of the sequence, returns list of factors of n
````

````python
# test codes
print(factorize(2))
print(factorize(72))
print(factorize(196))
````

2c. (10 points) Write a function that finds the prime factorization of a number.

````python
import math
def prime_factorize(n):
    #Finish this function
    #........
    answer = []
    while n % 2==0: #Checking how many factors of 2 n has 
        answer.append(2) #Adding factors of 2 to the list
        n=n/2 #Updating variable until condition fails
    for i in range(3,int(math.sqrt(n))+1,2): 
        while n % i == 0: #Finding factors other than 2
            answer.append(int(i)) #Adding factors to the list
            n=n/2 #Updating variable
    if n>2:
        answer.append(int(n))
    return answer
    
````

````python
# test codes
print(prime_factorize(2))
print(prime_factorize(72))
print(prime_factorize(196))
````

## 5. Recursive Functions
(20 points) The following function uses recursion to generate the $n$th row of Pascal's triangle:
```
              1
           1     1
        1     2     1
      1    3     3     1
   1    4     6     4    1
1     5    10    10    5    1
```

````python
def pascal(n):
    if n == 1:
        return [1]
    else:
        p_line = pascal(n-1)
        line = [ p_line[i]+p_line[i+1] for i in range(len(p_line)-1)]
        line.insert(0,1)
        line.append(1)
    return line

print(pascal(6))
````

Rewrite the above function to use iteration (using only `while` or `for` loops) instead of recursion.

````python
def pascal_2(n):
    n=n-1 #Looking to line above row
    line=[1] #Every row starts with integer 1
    for i in range(max(n,0)):
        line.append(int(line[i]*(n-i)/(i+1))) #Formula for row for Pascal triangle
    return line 


#test
pascal_2(6)
````
